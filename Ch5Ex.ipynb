## Problem 6 - Calorie counter
A nutritionist who works for a fitness club helps members by evaluating their diets. As part of her evaluation, she asks members for the number of fat grams and carbohydrate grams that they consumed in a day. Then, she calculates the number of calories that result from the fat, using the following formula:

calories from fat = fat grams × 9

Next, she calculates the number of calories that result from the carbohydrates, using the following formula:

calories from carbs = carb grams × 4

The nutritionist asks you to write a program that will make these calculations.

```python
# main function gets fat grams and carb grams
# from user and converts to calories from fat and
# carbs repectively
def main():
    print('This program will tell you the calories that\n'
         + 'result from fat based on your fat grams consumed\n'
         + 'and your calories from carbs based on your fat\n'
         + 'grams consumed. Enter your data below.')
    fat_grams = int(input('Fat grams: '))
    carb_grams = int(input('Carb grams: '))
    
    print(f'Your calories from fat are {calc_cal_fat(fat_grams):,.0f} calories.')
    print(f'Your calories from carbs are {calc_cal_carbs(carb_grams):,.0f} calories.')
    
# function calculates cals from fat
def calc_cal_fat(fat_grams):
    calories_from_fat = fat_grams * 9
    return calories_from_fat

# function calculates cals from carbs
def calc_cal_carbs(carb_grams):
    calories_from_carbs = carb_grams * 4
    return calories_from_carbs

# run main program
main()
```

## Problem 8 - Paint Job Estimator
A painting company has determined that for every 112 square feet of wall space, one gallon of paint and eight hours of labor will be required. The company charges $35.00 per hour for labor. Write a program that asks the user to enter the square feet of wall space to be painted and the price of the paint per gallon. The program should display the following data:

The number of gallons of paint required

The hours of labor required

The cost of the paint

The labor charges

The total cost of the paint job

```python
# Global constants for cost per hour
# and wall square footage factor
HOURLY_COST = 35
WALL_FACTOR = 112

# main program gets price of paint and square feet to paint
# displays gallons of paint needed, hours of labor, cost of paint,
# cost of labor, and total cost of job through calling multiple functions
def main():
    another_calc = input('Press \'y\' to start an estimation or\n'
                    + 'press \'n\' to end the program: ')
    
    while another_calc == 'y' or another_calc == 'Y':                      # while loop that iterates as many 
                                                                           # times as user wants
        paint_price = float(input('Enter the price of paint per gallon: '))
        square_feet = float(input('Enter the square footage to be painted: '))
    
        print(f'You need at least {paint_needed(square_feet):.0f} gallon(s) of paint\n'
             + 'to complete this job.\n')
    
        print(f'This job will take roughly {hours_to_finish(square_feet):.0f} hours to complete.\n')
          
        print(f'The total cost of paint is ${total_paint_cost(square_feet, paint_price):.2f}.\n')
          
        print(f'Costs of labor to complete the job come to ${labor_costs(square_feet, HOURLY_COST):.2f}.\n')
          
        print(f'Total cost of the job is ${total_cost(square_feet, paint_price, HOURLY_COST):.2f}\n')
    
        another_calc = input('Do you want to perform another job estimation? ')
          
# function that returns paint needed 
def paint_needed(square_feet):
    paint_needed = square_feet / WALL_FACTOR
    return paint_needed
    
# function that returns the hours to complete job
def hours_to_finish(square_feet):
    hours = (square_feet / WALL_FACTOR) * 8
    return hours
          
# function that returns the cost of paint
def total_paint_cost(square_feet, paint_price):
    total_paint_cost_value = (square_feet / WALL_FACTOR) * paint_price
    return total_paint_cost_value
    
# function that returns the labor costs
def labor_costs(square_feet, HOURLY_COST):
    labor_cost = (square_feet / WALL_FACTOR) * HOURLY_COST
    return labor_cost
          
# function that returns the total cost for the job 
def total_cost(square_feet, paint_price, HOURLY_COST):
    total_cost_value = total_paint_cost(square_feet, paint_price) + labor_costs(square_feet, HOURLY_COST)
    return total_cost_value

# call to main function
main()    
```

## Problem 9 - Monthly Sales Tax
A retail company must file a monthly sales tax report listing the total sales for the month, and the amount of state and county sales tax collected. The state sales tax rate is 5 percent and the county sales tax rate is 2.5 percent. Write a program that asks the user to enter the total sales for the month. From this figure, the application should calculate and display the following:

The amount of county sales tax

The amount of state sales tax

The total sales tax (county plus state)

```python
# main program prints intro message and
# gets total sales in month
# calculates county and sales tax
# adds the total and displays all 3
def main():
    print('This program calculates the county and state\n'
            + 'taxes on sales earnings in a month. Enter your\n'
            + 'total monthly sales below to begin.')
    print('-----------------------------------------------')
    num_calculations = int(input('How many calculations do you want to perform? '))
    print()
    
    for calculation in range(num_calculations):
        total_month_sales = float(input('Enter the amount of sales this month: $'))
        print()
        county_tax = calc_county_tax(total_month_sales)
        state_tax = calc_state_tax(total_month_sales)
        
        print(f'Calculation number {calculation + 1}: ', end='')
        print(f'County Tax Total: ${county_tax:,.2f}\n')
        print(f'Calculation number {calculation + 1}: ', end='')
        print(f'State Tax Total: ${state_tax:,.2f}\n')
        print(f'Calculation number {calculation + 1}: ', end='')
        print(f'Total Tax: ${county_tax + state_tax:,.2f}\n')
        print('-----------------------------------------------')
        
    
# function that calculates county tax
def calc_county_tax(monthly_sales):
    county_tax = monthly_sales * 0.025
    return county_tax


# function that calculates state tax
def calc_state_tax(monthly_sales):
    state_tax = monthly_sales * 0.05
    return state_tax


# call to main function
main()
```

## Problem 10 - Feet to Inches
One foot equals 12 inches. Write a function named feet_to_inches that accepts a number of feet as an argument and returns the number of inches in that many feet. Use the function in a program that prompts the user to enter a number of feet then displays the number of inches in that many feet.

```python
# main function prints intro and gets the 
# number of conversions to make from user
# calls feet_to_inches function to convert
# values to inches
def main():
    print('This program converts a number of feet\n'
         + 'to a number of inches.')
    print('**************************************')
    
    number_conversions = int(input('How many conversions do you need? '))
    
    for conversion in range(number_conversions):
        feet = float(input('Enter a number of feet: '))
        inches = feet_to_inches(feet)
        print(f'Conversion Number {conversion + 1}: ', end='')
        print(f'There are {inches:.1f} inches in {feet:.1f} feet.\n')
    
# function that converts feet to inches   
def feet_to_inches(number_feet):
    number_inches = number_feet * 12 
    return number_inches

# call to the main function 
main()
```

## Problem 13 - Falling Distance
When an object is falling because of gravity, the following formula can be used to determine the distance the object falls in a specific time period:

d = ½gt**2

The variables in the formula are as follows: d is the distance in meters, g is 9.8, and t is the amount of time, in seconds, that the object has been falling.

Write a function named falling_distance that accepts an object’s falling time (in seconds) as an argument. The function should return the distance, in meters, that the object has fallen during that time interval. Write a program that calls the function in a loop that passes the values 1 through 10 as arguments and displays the return value.

```python
# main function prints intro and prompts user
# to start. While loop iterates main program while 
# user wants to continue calculating falling time 
# in m/s based on user entered amount of seconds
def main():
    print('This program calculates falling distance\n'
         + 'in meters given an amount of seconds.')
    print('****************************************\n')
    
    print('Press the \'Enter\' key to start\n'
         + 'Press \'n\' to end the program.')
    
    another_calculation = input('')
    
    while another_calculation != 'n':
        falling_time = float(input('Enter the objects falling time in seconds: '))
        print(f'The item travelled {falling_distance(falling_time):,.1f} meters.')
        print('****************************************\n')
        
        another_calculation = input('Press the \'Enter\' key to perform another\n'
                                    + 'calculation. Press \'n\' to end the program.\n')
    
# function that calculates
# distance in meters 
def falling_distance(time):
    distance = 0.5 * 9.807 * time ** 2
    return distance

# call to main function
main()
```
