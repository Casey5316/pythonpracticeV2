CS101 8 a.m.

CS102 9 a.m.

CS103 10 a.m.

NT110 11 a.m.

The program should let the user enter a course number, then it should display the course’s room number, instructor, and meeting time.

```python
from ClassAdderMod import class_adder

# main function gets the course number from user and passes the course
# number to three functions to get the room number, teacher, and start time
# for the course entered, then displays all of the data.
def main():   
    ADD_CLASS = '1'
    FIND_COURSE_INFO = '2'
    
    room_numbers = {'CS101': '3004', 'CS102': '4501', 'CS103': '6755', 'NT110': '1244', 'CM241': '1411'}
    teachers = {'CS101': 'Haynes', 'CS102': 'Alvarado', 'CS103': 'Rich', 'NT110': 'Burke', 'CM241': 'Lee'}
    start_times = {'CS101': '8 AM', 'CS102': '9 AM', 'CS103': '10 AM', 'NT110': '11 AM', 'CM241': '1 PM'}

    
    
    selection = input('Enter \'1\' to add class information, or enter'
                     + '\'2\' to find an existing course: ')
    
    # calls the module function class_adder to add the key, value pairs
    # to the existing dictionaries
    if selection == '1':
        add_class = input('Would you like to add a class to the database? '
                         + 'Enter \'y\' for yes or \'n\' for no: ')

        while add_class.upper() == 'Y':
            new_course = input('Enter the new course number: ')
            print()

            new_room, new_teacher, new_time = class_adder()

            room_numbers[new_course] = new_room
            teachers[new_course] = new_teacher
            start_times[new_course] = new_time

            add_class = input('Would you like to add another class? '
                             + 'Enter \'y\' for yes or \'n\' for no: ')
        
    if selection == '2':
        course_number = input('Enter your course number: ')

        room_num = find_room_no(course_number, room_numbers)

        teacher = find_teacher(course_number, teachers)

        start = find_start_time(course_number, start_times)

        print(f'{course_number} starts at {start}, in Mr.\\Mrs. '
             + f'{teacher}\'s room: {room_num}')
            
            
# finds the room number from the dictionary of room numbers
# provided for each class. All functions are basically a copy
# and paste of eachother because they all do the same function 
# with different data 
def find_room_no(course, room_number_dict):
    for key in room_number_dict:
        if course in room_number_dict:
            value = room_number_dict.get(course)
            return value
        
    if course not in room_number_dict:
            print('Course not found')
        
    
    

# finds the teacher from the dictionary of teachers 
# provided for each class 
def find_teacher(course, teacher_dict):

    for key in teacher_dict:
        if course in teacher_dict:
            value = teacher_dict.get(course)
            return value
        
        
        
# finds the start time from the dictionary of start
# times provided for each class 
def find_start_time(course, start_time_dict):

    for key in start_time_dict:
        if course in start_time_dict:
            value = start_time_dict.get(course)
            return value        

        
    
            
            
            
            

# call to main function 
if __name__ == '__main__':
    main()
```
